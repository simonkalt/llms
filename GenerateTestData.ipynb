```python
# imports

import os
import requests
from IPython.display import Markdown, display, update_display
from openai import OpenAI
# from google.colab import drive
from huggingface_hub import login
# from google.colab import userdata
from transformers import AutoTokenizer, AutoModelForCausalLM, TextStreamer, BitsAndBytesConfig
import torch
```

```python
QWEN = "Qwen/Qwen2.5-Coder-32B-Instruct"
```

```python
# Sign in to HuggingFace Hub

hf_token = os.environ.get('HF_TOKEN')
login(hf_token, add_to_git_credential=True)
```

```python
system_message = "You are creating a list of the top 10 master categories and all related sub-categories.  For example, if the user chooses Automobiles (cars), you would return the Manufacturer as the master category, and all the models for each Manufacturer. "
user_prompt = f"User has requested the category of top 10 'Automobile Manufacturers by Sales'"

messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
  ]
```

```python
quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_type="nf4"
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(QWEN)
tokenizer.pad_token = tokenizer.eos_token
inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")
streamer = TextStreamer(tokenizer)
model = AutoModelForCausalLM.from_pretrained(QWEN, device_map="auto", quantization_config=quant_config)
outputs = model.generate(inputs, max_new_tokens=2000, streamer=streamer)
```
